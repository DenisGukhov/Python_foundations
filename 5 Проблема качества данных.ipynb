```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

Создадим датасет с бегунами на 100м\
Предположим, что результатаы будут лежать в диапозоне от 9.58(быстрее никто не бегал) до 12 секунд\
Введем также расу и возраст бегуна

```python
# Создаём сэмпл
n_samples = 1000
r = ['black', 'white', 'indian', 'asian']

runner_age = np.random.choice(range(17, 35), n_samples)
raсe = np.random.choice(r, n_samples)
height = np.random.choice(range(165, 205), n_samples)

result = np.random.choice(np.arange(9.58, 12, 0.01), n_samples)

data = pd.DataFrame({'runner_age': runner_age, 'raсe': raсe, 'height': height, 'result': result})
data.head(10)
```

Построим модель линейной регрессии и посмотрив веса признаков.\
Но сначала переведем расу в числа, чтобы модель работала

```python
from sklearn.metrics import mean_absolute_error

X = data[['runner_age', 'height']]
y = data['result']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['runner_age', 'height']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

Пронумеруем расы. Можно использовать лэйбл инкодер. Но у меня это есть в 11 домашке - поэтому просто скоппастим.

```python
data.raсe.replace(['black', 'white', 'indian', 'asian'], [0, 1, 2, 3], inplace=True)
```

```python
def get_score(X,y, random_seed=42, model=None):
  if model is None:
    model = LinearRegression()
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_seed )  
  model.fit(X_train, y_train)
  return model.score(X_test, y_test)
```
