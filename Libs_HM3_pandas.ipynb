Movies rating

```python
import pandas as pd
```

```python
raitings = pd.read_csv('C:/Projects/Netology/Homeworks/ratings.csv')
raitings.head()
```

```python
def grade_(x):
    if x['rating'] <= 2.0:
        return 'низкий рейтинг'
    elif 2.0 < x['rating'] <= 4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
raitings['class'] = raitings.apply(grade_, axis=1)
raitings.head()
```

GEO DATA

```python
keyword=pd.read_csv('C:/Projects/Netology/Homeworks/ml-latest-small/keywords.csv')
keyword
```

```python
def region(row):
    """Гео-классификатор на запросы"""
    if 'москва'in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'псков' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'undefined'

keyword['region'] = keyword.apply(region, axis=1)
keyword
```
