films grade and random data set

```python
import pandas as pd
```

```python
movies=pd.read_csv('C:/Projects/Netology/Homeworks/movies.csv')
movies[['movieId','title']].head()
```

```python
ratings = pd.read_csv('C:/Projects/Netology/Homeworks/ratings.csv')
ratings = ratings[['movieId','rating']]
```

```python
joined = ratings.merge(movies, on = 'movieId', how ='left')
joined[['movieId','rating','title']].head()
```

```python
max_ = ratings['rating'].max()
filtered_joined=joined[(joined ['rating'] == max_)]
fr = filtered_joined.groupby(['movieId' , 'title'])['movieId'].count()
```

```python
fr.sort_values(ascending=False).head()
```

```python
data=pd.read_csv('C:/Projects/Netology/Homeworks/power.csv')
data.head()
```

```python
baltic = data[(data['country'] == 'Lithuania')|(data['country'] =='Latvia')|(data['country'] =='Estonia')]
baltic.head(100)
```

```python
years_quantity = baltic[(baltic['year']>=2005) | (baltic['year'] <= 2010) | (baltic['quantity']>0)]
years_quantity.head()
```

```python
category = years_quantity[years_quantity.category.isin([4,12,21])]
category.head(100)
```
